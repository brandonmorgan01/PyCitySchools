```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

```python
#Set Variables for DataFrame Creation
number_schools = len(school_data_complete['school_name'].unique())
number_students = school_data_complete['student_name'].count()
total_budget = school_data_complete['budget'].unique().sum()
average_math = school_data_complete['math_score'].mean()
average_reading = school_data_complete['reading_score'].mean()

# number_passing_math = school_data_complete[school_data_complete.math_score > 69].count()
# percent_passing_math = (number_passing_math / number_students)

# number_passing_reading = school_data_complete[school_data_complete.reading_score > 69].count()
# percent_passing_reading = number_passing_reading / number_students

# overall_pass = (percent_passing_math + percent_passing_reading) / 2

```

```python
number_passing_math = school_data_complete[school_data_complete.math_score > 69].count()
percent_passing_math = (number_passing_math / number_students)

number_passing_reading = school_data_complete[school_data_complete.reading_score > 69].count()
percent_passing_reading = number_passing_reading / number_students

overall_pass = (percent_passing_math + percent_passing_reading) / 2
```

```python
#Create Summary Data Frame
district_df = pd.DataFrame({'Total Schools': num_schools, 
                                   'Total Students': num_students,
                                   'Total Budget': total_budget,
                                   'Average Math Score': average_math,
                                   'Average Reading Score': average_reading,
                                   '% Passing Math': percent_passing_math['math_score'],
                                   '% Passing Reading': percent_passing_reading['reading_score'],
                                   'Overall Passing Rate': overall_pass['Student ID']
                                   }, index = [0]
                                  )

#format
district_df.style.format({'Total Students': '{:,}',
                         'Total Budget': '${:,.2f}',
                         'Average Math Score':'{:.2f}',
                         'Average Reading Score':'{:.2f}',
                         '% Passing Math':'{:.2%}',
                         '% Passing Reading':'{:.2%}',
                         'Overall Passing Rate':'{:.2%}'
                         }
                        )
```

```python
#set index for school name and school type
school_type = school_data.set_index(["school_name"])["type"]

#set variables
tot_students = school_data_complete["school_name"].value_counts()

tot_budget = school_data.groupby(["school_name"]).sum()["budget"]

budget_by_student = tot_budget / tot_students

avg_math_score = school_data_complete.groupby(["school_name"]).mean()["math_score"]

avg_reading_score = school_data_complete.groupby(["school_name"]).mean()["reading_score"]

pct_pass_math = school_data_complete[(school_data_complete["math_score"] > 69)]
math_pass_list = pct_pass_math.groupby(["school_name"]).count()["student_name"] / tot_students 

pct_pass_reading = school_data_complete[(school_data_complete["reading_score"] > 69)]
reading_pass_list = pct_pass_reading.groupby(["school_name"]).count()["student_name"] / tot_students 

overall_pass_rate = (math_pass_list + reading_pass_list) / 2




# avg_math_df = pd.DataFrame(school_group["math_score"].mean())
# avg_reading_df = pd.DataFrame(school_group['reading_score'].mean())
# avg_reading_df
```

```python
#create data fram summary by school
school_summary = pd.DataFrame({"School Type": school_type,
                                "Total Students": tot_students,
                                "Total School Budget": tot_budget,
                                "Per Student Budget": budget_by_student,
                                "Average Math Score": avg_math_score,
                                "Average Reading Score": avg_reading_score,
                                "% Passing Math": math_pass_list,
                                "% Passing Reading": reading_pass_list,
                                "Overall Passing Rate": overall_pass_rate
                              }
                             )

#format
school_summary_formatted = school_summary.style.format({"Total Students":'{:,}',
                                                        "Total School Budget":'${:,.2f}',
                                                        "Per Student Budget":'${:,.2f}',
                                                        "Average Math Score":'{:.2f}',
                                                        "Average Reading Score":'{:.2f}',
                                                        "% Passing Math":'{:.2%}',
                                                        "% Passing Reading":'{:.2%}',
                                                        "Overall Passing Rate":'{:.2%}',
                                                       }
                                                      )
school_summary_formatted
```
